```python
df = df_UCL
df.columns = [ int(x.replace('年','')) if x!='地区' else x for x in df.columns]
years = [x for x in df.columns if x!='地区']
df_melted = pd.melt(df, id_vars=['地区'],value_vars = years)
df_melted.columns = ['地区','年','居民消费水平']
df_UCL_done = df_melted
df_UCL_done
```

```python
df_p = df_UCL_done.join(df_GDP_done,lsuffix='',rsuffix='_GDP')
df_p
```

```python
df_p.drop(["地区_GDP","年_GDP"],axis=1)
```

```python
df_p.drop(["地区_GDP","年_GDP"],axis=1).set_index(["地区","年"])
```

```python
df_pp = df_p.drop(["地区_GDP","年_GDP"], axis=1).set_index(["地区","年"])
```

```python
df_pp.query("年==2015")
```

```python
%matplotlib inline
df_pp.query("年==2015").plot(kind='scatter',x='居民消费水平',y='GDP',
          title = '居民消费水平与GDP的关系：分省数据')
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import pylab
mpl.rcParams['font.sans-serif']=['SimHei']
mpl.rcParams['axes.unicode_minus']=False

df_plot = df_pp.query("年==2015")
df_plot.plot(kind='scatter',x='居民消费水平',y='GDP',
            title = '居民消费水平与GDP的关系：分省数据')
```

```python
import numpy as np
fit = np.polyfit(x=df_pp.query("年==2015")['居民消费水平'],y=df_pp.query("年==2015")['GDP'],deg=1)
```

```python
x=df_plot['居民消费水平']
y=df_plot['GDP']
ax = df_plot.plot(kind='scatter', x='居民消费水平', y='GDP', 
          title = '居民消费水平与GDP的关系：分省数据')
ax.plot(x, fit[0] * x + fit[1], color='red')
ax.scatter(x, y)
```
